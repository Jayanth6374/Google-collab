```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset (replace this with your actual dataset)
data = {
    'User': ['User1', 'User2', 'User3', 'User4', 'User5'],
    'Query': ['IssueA', 'IssueB', 'IssueA', 'IssueC', 'IssueB'],
    'ResolutionTime': [10, 15, 8, 12, 18],
    'Feedback': ['Positive', 'Negative', 'Positive', 'Neutral', 'Positive']
}

df = pd.DataFrame(data)

# Descriptive statistics
print("Descriptive Statistics:")
print(df.describe())

# Exploratory Data Analytics
# 1. Frequency distribution of user queries
query_distribution = df['Query'].value_counts()

# 2. Average response time for different types of queries
avg_response_time = df.groupby('Query')['ResolutionTime'].mean()

# 3. User feedback sentiments distribution
feedback_distribution = df['Feedback'].value_counts()

# Visualizations
# 1. Pie chart for the distribution of user queries
plt.figure(figsize=(8, 8))
plt.pie(query_distribution, labels=query_distribution.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of User Queries')
plt.show()

# 2. Bar chart for average response time
avg_response_time.plot(kind='bar', color='skyblue')
plt.title('Average Response Time for Different Queries')
plt.xlabel('Query')
plt.ylabel('Average Response Time')
plt.show()

# 3. Pie chart for user feedback sentiments
plt.figure(figsize=(8, 8))
plt.pie(feedback_distribution, labels=feedback_distribution.index, autopct='%1.1f%%', startangle=90)
plt.title('User Feedback Sentiments Distribution')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset (replace this with your actual dataset)
data = {
    'User': ['User1', 'User2', 'User3', 'User4', 'User5'],
    'Country': ['USA', 'Canada', 'USA', 'India', 'UK'],
    'UsageFrequency': [5, 10, 3, 8, 12],
    'SatisfactionRating': [4, 3, 5, 2, 4],
    'IssueResolved': ['Yes', 'No', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)

# Univariate Analysis
# 1. Global usage frequency distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['UsageFrequency'], kde=True)
plt.title('Global Chatbot Usage Frequency Distribution')
plt.xlabel('Usage Frequency')
plt.ylabel('Count')
plt.show()

# 2. Country-wise chatbot users
country_wise_users = df['Country'].value_counts()
plt.figure(figsize=(12, 6))
country_wise_users.plot(kind='bar', color='skyblue')
plt.title('Country-wise Chatbot Users')
plt.xlabel('Country')
plt.ylabel('Number of Users')
plt.show()

# Bivariate Analysis
# 1. Scatter plot: Usage frequency vs. Satisfaction rating
plt.figure(figsize=(10, 6))
sns.scatterplot(x='UsageFrequency', y='SatisfactionRating', data=df)
plt.title('Scatter Plot: Usage Frequency vs. Satisfaction Rating')
plt.xlabel('Usage Frequency')
plt.ylabel('Satisfaction Rating')
plt.show()

# 2. Box plot: Issue resolved vs. Satisfaction rating
plt.figure(figsize=(10, 6))
sns.boxplot(x='IssueResolved', y='SatisfactionRating', data=df)
plt.title('Box Plot: Issue Resolved vs. Satisfaction Rating')
plt.xlabel('Issue Resolved')
plt.ylabel('Satisfaction Rating')
plt.show()

# Multivariate Analysis
# 1. Correlation matrix
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset (replace this with your actual dataset)
data = {
    'User': ['User1', 'User2', 'User3', 'User4', 'User5'],
    'Country': ['USA', 'Canada', 'USA', 'India', 'UK'],
    'UsageFrequency': [5, 10, 3, 8, 12],
    'SatisfactionRating': [4, 3, 5, 2, 4],
    'IssueResolved': ['Yes', 'No', 'Yes', 'No', 'Yes']
}

df = pd.DataFrame(data)

# Univariate Analysis
# 1. Global usage frequency distribution
print("Global Chatbot Usage Frequency Distribution:")
print(df['UsageFrequency'].describe())

plt.figure(figsize=(10, 6))
sns.histplot(df['UsageFrequency'], kde=True)
plt.title('Global Chatbot Usage Frequency Distribution')
plt.xlabel('Usage Frequency')
plt.ylabel('Count')
plt.show()

# 2. Country-wise chatbot users
country_wise_users = df['Country'].value_counts()
print("\nCountry-wise Chatbot Users:")
print(country_wise_users)

plt.figure(figsize=(12, 6))
country_wise_users.plot(kind='bar', color='skyblue')
plt.title('Country-wise Chatbot Users')
plt.xlabel('Country')
plt.ylabel('Number of Users')
plt.show()

# Bivariate Analysis
# 1. Scatter plot: Usage frequency vs. Satisfaction rating
print("\nBivariate Analysis - Scatter Plot: Usage Frequency vs. Satisfaction Rating:")
print(df[['UsageFrequency', 'SatisfactionRating']].corr())

plt.figure(figsize=(10, 6))
sns.scatterplot(x='UsageFrequency', y='SatisfactionRating', data=df)
plt.title('Scatter Plot: Usage Frequency vs. Satisfaction Rating')
plt.xlabel('Usage Frequency')
plt.ylabel('Satisfaction Rating')
plt.show()

# 2. Box plot: Issue resolved vs. Satisfaction rating
print("\nBivariate Analysis - Box Plot: Issue Resolved vs. Satisfaction Rating:")
print(df.groupby('IssueResolved')['SatisfactionRating'].describe())

plt.figure(figsize=(10, 6))
sns.boxplot(x='IssueResolved', y='SatisfactionRating', data=df)
plt.title('Box Plot: Issue Resolved vs. Satisfaction Rating')
plt.xlabel('Issue Resolved')
plt.ylabel('Satisfaction Rating')
plt.show()

# Multivariate Analysis
# 1. Correlation matrix
print("\nMultivariate Analysis - Correlation Matrix:")
correlation_matrix = df.corr()
print(correlation_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
